```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
boston_data = load_boston()
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
```

```python
df.columns
```

```python
df.head()
```

# Standardization / Mean Removal / Variance Scaling

```python
from sklearn import preprocessing
```

```python
df_std = preprocessing.StandardScaler().fit_transform(df)
pd.DataFrame(df_std, columns=boston_data.feature_names).head()
```

# Min-Max or Scaling Features to a Range

```python
df_minmax = preprocessing.MinMaxScaler().fit_transform(df)
pd.DataFrame(df_minmax, columns=boston_data.feature_names).head()
```

# Normalization

```python
df_norm = preprocessing.Normalizer().fit_transform(df)
pd.DataFrame(df_norm, columns=boston_data.feature_names).head()
```
